Importando a biblioteca pandas que será utilizada para a leitura dos dados.

```
import pandas as pd
```

1. Importação dos dados de gastos Hospitalares do SUS por região e unidade federativa.

1.1 Importação do arquivo csv com o pandas. 

Na importação tem que ser realizado os seguintes tratamentos:

*   Ignorar as linhas iniciais que não farão parte dos dados a serem explorados.
*   Ignorar as linhas finais que não farão parte dos dados a serem explorados.
*   Especificar o carater delimitador que no nosso caso é o ';'.
*   O encoding para ser compatível com a lingua portuguesa.
*   Indicar qual o caracter utilizado para milhar e dezena para que o panda formate os valores monetários corretamente.







```

dados = pd.read_csv('https://raw.githubusercontent.com/giziellerodrigues/agendamento-hospitalar/main/dados/A151346189_28_143_208.csv', encoding='ISO-8859-1', skiprows=3, sep=';', skipfooter=12, thousands=".", decimal=",")
dados=dados.replace(to_replace="-",value=0)
dados
```

Exibindo os 5 primeiros registros




```
dados.head()
```

Exibi os últimos registros

```
dados.tail()
```

Obtendo a média

```
 dados.mean()
```

Obtendo as informações da composição dos dados

```
dados.info()
```

Configurando a opção global do pandas para formatação numérica

```
pd.options.display.float_format = "{:.2f}".format
```

Média com os números formatados

```
dados.mean()
```

Obtendo os valores dos gastos hospitalares de agosto de 2008

```
dados["2008/Ago"]
```

Gerando a média dos gastos hospitalares de agosto de 2008

```
dados["2008/Ago"].mean()
```

Gerando um grafico com os gastos por estados de agosto de 2008

```
dados.plot(x="Unidade da Federação", y="2008/Ago")
```

Alterando o tipo do gráfico para melhorar a exibição das informações e torná-las mais coerentes, e aumentando o tamanho para melhor visualização da informação.

```
dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
```

Alterando configurações mais avançadas do gráfico com a biblioteca matplotlib

```
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker


ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title('Valor por unidade da federação')
plt.show()
```

Desafio 1 -> escolher um título mais descritivo, que passe a mensagem adequada

```
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker


ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6), label='Valor gasto em reais')
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}"))

plt.title('Valor Despesas hospitalares por unidade da federação em Agosto de 2008')
plt.show()
```

Desafio 2 -> Análise dos gastos hospitalares para o mês de agosto de 2020


```
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

dados["2020/Ago"] = pd.to_numeric(dados["2020/Ago"])

ax = dados.plot(x="Unidade da Federação", y="2020/Ago", kind="bar", figsize=(9,6), label='Valor gasto em reais')
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}"))

plt.title('Valor Despesas hospitalares por unidade da federação em Agosto de 2020')
plt.show()
```
